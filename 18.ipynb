```python
s = [-1,0,1,2,-1,-4]
target = -1
s.sort()
out = []
for i in range(len(s)-3) :
    if i > 0 and s[i] == s[i-1] : continue
    for j in range(i+1,len(s)-2) :
        if j > i + 1 and s[j] == s[j-1] : continue
        head = j + 1
        tail = len(s) -1
        temp_target = target - (s[i]+s[j])
        while head < tail :
            temp = s[head] + s[tail]
            if temp == temp_target :
                out.append([s[i],s[j],s[head],s[tail]])
                while head < tail and s[head] == s[head+1] : head += 1
                while head < tail and s[tail] == s[tail-1] : tail -= 1
                head += 1
                tail -= 1
            elif temp > temp_target : tail -= 1
            else : head += 1
print out
```
