```python
roman = { 'I' : 1 , 'V' : 5 , 'X' : 10 , 'L' : 50 , 'C' : 100 , 'D' : 500 , 'M' : 1000 }
x = 'XVIII'
out = 0
for i in range(len(x)) :
    if i != len(x)-1 and roman[x[i]] < roman[x[i+1]]   :
        temp = -roman[x[i]]
    else :
        temp = roman[x[i]]
    out = out + temp
print out
```
