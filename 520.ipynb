```python
word = "AAhA"
out = True
if len(word) == 1 : print True
if ord(word[0]) >= ord('a') :
    head = 0
elif ord(word[1]) >= ord('a') :
    head = 0
else : 
    head = 1
for i in range(1,len(word)) :
    if ord(word[i]) < ord('a') and head == 0 :
        out = False
        break
    elif ord(word[i]) >= ord('a') and head == 1 :
        out = False
        break
print out
```
