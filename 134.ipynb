```python
gas  = [1,2,3,4,5]
cost = [3,4,5,1,2]
```

```python
output = -1
for i in range(len(gas)) :
    temp , start = gas[i] , i
    while True : 
        add = start+1 if start+1 < len(gas) else 0
        temp -= cost[start] 
        if temp >= 0 :
            temp += gas[add]
        else :
            break
        start = start + 1
        if start == len(gas) :
            start = 0
        if start == i :
            break
    if temp > 0 :
        output = i
print output
```
