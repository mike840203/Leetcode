```python
# 普通遞迴
def fabona(index) :
    if index <= 1 :
        return 1
    return fabona(index-1) + fabona(index-2)
print fabona(10)
```

```python
# 普通DP
def ff(index) :
    fib = [0] * (index+1)
    fib[0] = fib[1] = 1
    for i in range(2,index+1) :
        fib[i] = fib[i-1] + fib[i-2]
    return fib[index]
ff(10)
```

```python
# 記憶行遞迴 = DP
def f(index) :
    if index == 1  : 
        return 1
    if index == 0 : 
        return 0
    if fib[index] != 0 :
        return fib[index]
    fib[index] = f(index-1) + f(index-2)
    return fib[index]


index = 10
fib = [0] * ( index + 1 )
print f(index) , fib

   
```
