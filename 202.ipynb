```python
n = 0
def isHappy(num,times=0) :
    times = times + 1
    temp = 0
    num = str(num)
    for i in range(len(num)) :
        temp = temp + int(num[i]) * int(num[i])
    if temp == 1 :
        return True 
    if times > 100 :
        return False
    else : 
        return isHappy(temp,times)
isHappy(n)
```
