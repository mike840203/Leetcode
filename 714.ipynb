```python
'''
There are two state for this problem 

state 1 (buy) : 
    do nothing --> state 1 
    sell --> state 2
state 2 (sell) : 
    do nothing --> state 2 
    buy --> state 1
    
buy , sell = -prices[0] , float('-inf')
Transaction comes from the sell state
'''
prices = [9,8,7,1,2]
fee = 2
buy , sell = -prices[0] , 0
for i in prices :
    buy , sell = max(buy,sell-i) , max(sell,buy+i-fee)
print sell
```
