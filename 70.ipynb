```python
n = 6
num1 = n % 2
num2 = n / 2
out = 0
print 'num1 = ' + str(num1)
print 'num2 = ' + str(num2)
def floor(num) :
    temp = 1
    for i in range(1,num+1):
        temp = temp * i
    return temp 
def C(upper,lower) :
    if upper >= lower :
        return floor(upper)/(floor(lower)*floor(upper-lower))
    else :
        return 0
for i in range(1,num2+1) :
    num1 = n - 2 * i
    out = out + C(num1+i,i)
    print 'out = ' + str(out)
print out + 1
```
