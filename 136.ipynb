```python
import timeit
s = timeit.default_timer()
nums = [1,1,2,2,5,5,8,7,9,6,6,8,7]
for i in range(len(nums)) :
    for j in range(i,len(nums)) :
        if nums[i] > nums[j] :
            temp = nums[i]
            nums[i] = nums[j]
            nums[j] = temp
t1 = timeit.default_timer()
res = 0
for i in nums :
    res ^= i
t2 = timeit.default_timer()
print res
print t1 - s
print t2 - t1
```
