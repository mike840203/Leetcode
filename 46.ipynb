```python
def permute(nums):
    return [[n] + p
            for i, n in enumerate(nums)
            for p in permute(nums[:i] + nums[i+1:])] or [[]]
permute(nums)
```

```python
def dfs(nums,path,res) :
    if not nums :
        res.append(path)
    for i in range(len(nums)) :
        dfs(nums[:i]+nums[i+1:],path+[nums[i]],res)
def permute(nums) :
    res = []
    dfs(nums,[],res)
    return res
nums = [1, 2, 3]
permute(nums)
```
