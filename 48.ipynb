```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
print id(matrix)
matrix.reverse()
print id(matrix)
matrix = matrix[::-1]
print id(matrix)
length = len(matrix)
for i in range(0,length-1) :
    for j in range(i+1,length) :
        temp = matrix[i][j]
        matrix[i][j] = matrix[j][i]
        matrix[j][i] = temp
print matrix
```
