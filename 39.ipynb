```python
candidates = [2, 3, 6, 7]
target = 7
def check(candidates,target,index,path,res) :
    if target == 0 : 
        res.append(path)
        return
    for i in range(index,len(candidates)) :
        if candidates[i] > target :
            break
        check(candidates,target-candidates[i],i,path+[candidates[i]],res)
def combinationSum(candidates , target):
    candidates.sort()
    out = []
    check(candidates,target,0,[],out)
    return out

combinationSum(candidates,target)
            
```
