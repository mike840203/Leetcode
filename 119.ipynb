```python
k = 0
i = 1
out = []
if out == [] :
    out.append(1)
while i <= k :
    temp = [a for a in out[i-1]]
    for x in range(1,i+1) :
        if x == i :
            temp.append(1)
        else :
            temp[x] = out[i-1][x-1] + out[i-1][x]
    out.append(temp)
    i = i + 1
print out[k]
            
```
