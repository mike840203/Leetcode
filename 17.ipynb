```python
digit = "279"
def cal(digits,output) : 
    if len(digits) == 0 :
        return output
    else :
        temp = []
        num = int(digits[0])
        head = 91 + num * 3 + ( 1 if num > 7 else 0 )
        offset = 4 if num == 9 or num == 7 else 3
        if len(output) == 0 :
            for i in range(head,head+offset) :
                temp.append(chr(i))
        else :
            for i in output :
                for j in range(head,head+offset) :
                    temp.append(i+chr(j))
        return cal(digits[1::],temp)
print cal(digit,[])
```
