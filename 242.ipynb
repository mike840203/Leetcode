```python
s = "anagram"
t = "nagaram"
sdict = {}
if len(s) ==0 and len(t) == 0 : print True
if len(s) != len(t) : print False
for element in s :
    if element not in sdict :
        sdict[element] = 1
    else :
        sdict[element] += 1
for element in t :
    if element not in sdict :
        print False
    else :
        sdict[element] -= 1
    if sdict[element] < 0 :
        print False
    
```
