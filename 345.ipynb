```python
s = "aA"
s = [i for i in s]
head = 0
tail = len(s) - 1
sh = 0
st = 0
listx = ['a','e','i','o','u','A','E','I','O','U']
while head < tail :
    if s[head] in listx :
        sh = 1
    else :
        head += 1
    if s[tail] in listx :
        st = 1
    else :
        tail -= 1
    if sh == 1 and st == 1 :
        temp = s[head]
        s[head] = s[tail]
        s[tail] = temp
        sh = 0
        st = 0
        head += 1
        tail -= 1

print "".join(s)
    
```
