```python
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
if len(strs) == 0 : print []
outlist = [[strs[0]]]
for element in strs[1:] :
    for row in range(len(outlist)) :
        if sorted(element) == sorted(outlist[row][0]) :
            outlist[row].append(element)
            break
        elif row == len(outlist) - 1 :
            outlist.append([element])
print outlist
```

```python
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
dict = {}
for element in strs :
    temp = "".join(sorted(element))
    if temp in dict.keys() :
        dict[temp].append(element)
    else :
        dict[temp] = [element]
print [dict[x] for x in dict.keys()]
```
