```python
s = "abctgtgtrgrt"
out = 0
for i in range(len(s)) :
    for j in range(len(s)) :
        if i + j < len(s) and s[i:i+j+1] == s[i:i+j+1][::-1] :
            out += 1
print out
```

```python
# 中心擴散法
s = "abctgtgtrgrt"
out = 0
def helper(s,left,right) :
    temp = 0
    while left >= 0 and right < len(s) and s[left] == s[right] :
        left -= 1
        right += 1
        temp += 1
    return temp
for i in range(len(s)) :
    out += helper(s,i,i)
    out += helper(s,i,i+1)
print out
```

```python
# DP 1-5回文 = 2-4回文
s = "abctgtgtrgrt"
dp = [[False]*len(s)]*len(s)
out = 0
for i in range(len(s)) :
    for j in range(i+1) :
        dp[i][j] = (s[i] == s[j]) and ( i - j <= 2 or dp[i-1][j+1] )
        if dp[i][j] :
            out +=1
print out
```
