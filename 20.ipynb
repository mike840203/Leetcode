```python
x = "[({(())}[()])]"
kinds = {'(' : 1 , '[' : 2 , '{' :3 , ')' : -1 , ']' : -2 , '}' : -3 }
justice = False
length = len(x)
if len(x) % 2 == 0 :
    for i in range(0,length-1) :
        if kinds[x[i]] + kinds[x[i+1]] == 0 or kinds[x[i]] + kinds[x[length-1-i]] == 0 :
            justice = True
        elif kinds[x[i]] > 0 :
            default = 0
            for j in range(i+1,length) : 
                if kinds[x[i]] == kinds[x[j]] :
                    default = default + 1
                elif kinds[x[i]] == -kinds[x[j]] :
                    default = default - 1
                else :
                    pass
                if default == - 1 :
                    default = 0
                    temp = 0
                    for k in range(i,j+1) :
                        temp = temp + kinds[x[k]]
                    if temp == 0 :
                        justice = True
                        break
                    else :
                        justice = False
                        break
                elif j == length - 1 and default != -1:
                    justice = False
            if justice == False :
                break
print justice
```
