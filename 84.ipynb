```python
heights = [2,1,5,6,2,3]
temp = 0
for i in range(len(heights)) :
    if heights[i] != 0 :
        for high in range(1,heights[i]+1) :
            length = 1
            for k in range(i+1,len(heights)) :
                if heights[k] >= high :
                    length = length + 1
                else :
                    break
            temp = max(temp,high*length)
print temp
```

```python
height = [2,1,5,0,6,2,3]
heights = list(set(height))
out = 0
for i in range(len(heights)) :
    ma , conti = 0 , 0
    for j in range(len(height)) :
        if heights[i] > height[j] :
            conti = 0
        else :
            conti += 1
        ma = max(ma,conti)
    out = max(out,heights[i]*ma)
print out
```

```python
def largestRectangleArea(height):
    height.append(0)
    stack = [-1]
    ans = 0
    for i in xrange(len(height)):
        while height[i] < height[stack[-1]]:
            h = height[stack.pop()]
            w = i - stack[-1] - 1
            ans = max(ans, h * w)
            print i ,h , w , ans , stack
        stack.append(i)
    height.pop()
    return ans
```
