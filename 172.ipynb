```python
def scale_recurr(n) :
    if n == 0: return 1
    if n >= 1 :
        return n * scale_recurr(n-1)
    else :
        return n
scale_recurr(5)
```

```python
def scale_for(n) :
    out = 1
    if n == 0 : out = 1
    for i in range(1,n+1) :
        out = out * i
    return out
scale_for(0)
```

```python
def getNumOf5InN(n) :
    if n < 5 : return 0 
    out = 0
    while n >= 5 :
        out = out + n / 5
        n = n / 5
    return out
getNumOf5InN(25)
```
