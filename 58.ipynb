```python
s = "Today is a nice day"
spane = []
for i in range(len(s)) :
    if s[i] == ' ' :
        spane.append(i)
if len(s) == len(spane):   #全空
    print 0
elif spane == [] :         #沒空
    print len(s)
else :
    head = spane[0]
    tail = spane[len(spane)-1]
    if head == 0 and tail == len(s) - 1 :
        for i in range(len(spane)-1,-1,-1) :
            if spane[i] != spane[i-1] + 1 :
                print spane[i] - spane[i-1] - 1 
                break
    elif head == 0 :
        print len(s) - tail - 1 
    elif tail == len(s) - 1 :
        for i in range(len(spane)-1,-1,-1) :
            if i == 0 :
                print spane[0]
                break
            if spane[i] != spane[i-1] + 1 :
                print spane[i] - spane[i-1] - 1 
                break    
    else :
        print len(s) - spane[len(spane)-1] - 1
print spane
```
