```python
matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]
iindex , jindex = [] , []
for i in range(len(matrix)) :
    for j in range(len(matrix[0])) :
        if matrix[i][j] == 0 :
            iindex.append(i)
            jindex.append(j)
print iindex , jindex
for element in iindex :
    for j in range(len(matrix[0])) :
        matrix[element][j] = 0
for element in jindex :
    for i in range(len(matrix)) :
        matrix[i][element] = 0   
print matrix
```
