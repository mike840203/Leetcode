```python
from collections import Counter
s = "baa"
p = "aa"
out = []
dictp = Counter(p)
dicts = Counter(s[:len(p)-1])
for i in range(len(p)-1,len(s)) :
    dicts[s[i]] += 1
    if dicts == dictp :
        out.append(i-len(p)+1)
    dicts[s[i-len(p)+1]] -= 1 
    if dicts[s[i-len(p)+1]] ==0 :
        del dicts[s[i-len(p)+1]]
print out
```
