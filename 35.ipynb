```python
nums = [1,3,5,6]
target = 8
for x in range(len(nums)) :
    if nums[x] == target :
        print x
    elif nums[x] >= target and x == 0 :
        print x
    elif nums[x] <= target and x == len(nums) - 1 :
        print x + 1
    elif nums[x] < target and target < nums[x+1] :
        print x + 1
    else :
        pass
        
```
