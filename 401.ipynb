```python
num = 9
if num == 0 : print ["0:00"]
if num > 8  : print []
def countbit(num) :
    listx = [32,16,8,4,2,1]
    count = 0
    for i in listx :
        if i > num :
            pass
        elif i <= num :
            num = num - i
            count = count + 1
            if num == 0 : break
    return count
out = []
for i in range(0,12) :
    for j in range(0,60) :
        if countbit(i) + countbit(j) == num :
            temp = str( i ) + ( ':0' if j < 10 else ':' ) + str( j )
            out.append(temp)
print out
```
