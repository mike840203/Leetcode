```python
def cal( num , head = 0 , tail = 0 ) :
    if tail == 0 : tail = num
    thisnum = ( head + tail + 1 ) / 2
    if num == 1 : return True
    if head + 1 == tail :
        return False
    if thisnum * thisnum > num :
        return cal(num,head,thisnum)
    elif thisnum * thisnum < num :
        return cal(num,thisnum,tail)
    else :
        return True
cal(1)
```
