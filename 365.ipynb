```python
x = 1
y = 1
z = 12
if z == 0 : print True
if x == 0 and y == 0 : print False
if x + y < z : print False
common = 0
mi , ma = min(x,y) , max(x,y) 
if mi == 0 : 
    mi = ma
for i in range(mi,0,-1) :
    if ma % i == 0 and mi % i == 0:
        common = i
        break
if z % ma == 0 or z % mi == 0 : print True
devider = min(x,y) if min(x,y) != 0 else ma
print common
print devider
print ( z % devider ) % common == 0
```
