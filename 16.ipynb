```python
s = [-1,0,1,1,55]
s.sort()
target = 3
temp = 0
min = 1000
for i in range(len(s)-2) :
    if i > 0 and s[i] == s[i-1] :continue
    head = i + 1
    tail = len(s) - 1
    while head < tail :
        sum = s[i] + s[head] + s[tail]
        if min > abs(target-sum) :
            min = abs(target-sum)
            temp = sum
        if target-sum > 0 : head += 1
        else : tail -= 1      
print temp
```
