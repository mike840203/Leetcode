```python
import timeit
numbers=[2,2,2,2,5, 7, 11, 15]
target= 18
out = []
s = timeit.default_timer()
for i in range(len(numbers)) :
    if i == 0 :
        tempx = numbers[0]
    if numbers[i] != tempx or i == 0 :
        for j in range(i+1,len(numbers)) :
            if j ==  i + 1 :
                tempy = numbers[j]
            if (numbers[j] != tempy or j == i + 1) and numbers[i] + numbers[j] == target   :
                out.append(i+1)
                out.append(j+1)
                break
print out
t = timeit.default_timer()
print t-s
```
