```python
haystack = "mississippi"
needle = "pi"
out = -1
if len(haystack) == 0 and len(needle) == 0 :
    out = 0
elif len(haystack) > len(needle) :
    for i in range(len(haystack)-len(needle)+1) :
        if haystack[i:i+len(needle)] == needle :
            out = i
            break
elif len(haystack) == len(needle) :
    if haystack == needle :
        out = 0
else :
    pass
print out
```
