```python
nums = [2,0,2,1,1,0,2,1,0,2]
c1 , c2 , index = 0 , len(nums)-1 , 0 
while index <= c2 :
    if nums[index] == 0 :
        nums[index] , nums[c1] = nums[c1] , nums[index]
        c1 += 1
    if nums[index] == 2 :
        nums[index] , nums[c2] = nums[c2] , nums[index]
        c2 -= 1
        index -= 1
    index += 1
print nums
```
