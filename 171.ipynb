```python
s = 'AAA'
out = 0
def cal(s,out,pw) :
    if len(s) > 1 :
        out = out + (ord(s[::-1][0]) -64)*pow(26,pw)
        return cal(s[:len(s)-1],out,pw+1)
    else :
        out = out + (ord(s) - 64)*pow(26,pw)
        return out
print cal(s,out,0)
```
