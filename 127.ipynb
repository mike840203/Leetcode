```python
begin = "hit"
end = "cog"
word = ["hot","dot","dog","lot","log","cog"]
```

```python
def compare(A,B) :
    count = 0
    for i in range(len(A)) :
        if A[i] != B[i] :
            if count == 1 :
                return False
            count += 1
    return True
```

```python
def bfs(start,end,list,index) :
    print start , end , list , index
    temp = 0
    if end == start :
        print "Complete"
        return index
    count = 0
    for i in range(len(list)) :
        if compare(start,list[i]) :
            if bfs(list[i],end,list[0:i]+list[i+1:],index+1) != 0 :
                temp = min(temp,bfs(list[i],end,list[0:i]+list[i+1:],index+1))
            count = 1
    if count == 0 :
        print "No avalible"
        return 0
    return temp       
bfs(begin,end,word,0) 
```
