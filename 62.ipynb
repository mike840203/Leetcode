```python
def uniquePaths(m, n):
    out = []
    cal(m,n,0,0,out)
    print sum(out)
def cal(right,down,index_r,index_d,res) :
    if index_r == right and index_d == down :
        res.append(1)
        return
    for r in range(index_r,right+1) :
        for d in range(index_d,down+1) :
            if d >= down or r >= right :
                break
            cal(right,down,r+1,d+1,res)
uniquePaths(7,3)
```

```python
def ccc(m,n) :
    if m <= 0 or n <= 0 : return 0
    if m == 1 and n == 1 : return 1
    return ccc(m-1,n) + ccc(m,n-1)
print ccc(23,12)
```

```python
def xxx(m,n) :
    mat = [[0 for i in range(m)]for j in range(n)]
    for y in range(n) :
        for x in range(m) :
            if x == 0 and y == 0 : 
                mat[y][x] = 1
            else :
                mat[y][x] = mat[y-1][x] + mat[y][x-1]
    return mat[y][x]
print xxx(12,10)
```
