```python
nums = [1]
target = 0

def search(nums,target) :
    head , tail = 0 , len(nums) - 1 
    while head + 1 < tail :
        n = ( head  + tail ) / 2 
        if nums[n] > target :
            tail = n
        elif nums[n] < target :
            head = n
    return tail

def searchRange(nums, target):
    if len(nums) == 0 : print [-1,-1] 
    nums = [-1000] + nums
    nums.append(1000)
    head = search(nums,target-0.5) 
    tail = search(nums,target+0.5) - 1 
    head = head if nums[head] == target else -1
    tail = tail if nums[tail] == target else -1
    if head == -1 :
        print [head,tail]
    else :
        print [head-1,tail-1]
searchRange(nums,target)
    
```
