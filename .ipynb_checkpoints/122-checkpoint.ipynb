```python
prices = [7, 1, 5, 100, 6, 1000 , 1 , 50 , 2 , 40]
m = 0
ma = -1
mi = 1000
for i in range(len(prices)-1) :
    if prices[i+1] > prices[i] and prices[i+1] > ma :
        ma = prices[i+1]
        if mi > prices[i] :
            mi = prices[i]
        m = max(m,ma-mi)
    elif prices[i+1] < mi :
        mi = prices[i+1] 
        ma = -1
print m
# 如果右邊>左邊 和 右邊>最大值 的話
  #  最大值 = 右邊
  #  如果最小值>左邊
     # 最小值 = 左邊
# 否則如果 最小值 > 右邊
  #  最小值 = 右邊
  #  最大值 = -1
```
