```python
n = 4
matrix = [[0 for i in range(n)] for i in range(n)]
print matrix
up , right , down , left = 0 , 0 , 0 , 0
length , height = len(matrix[0]) , len(matrix)
output , count = [] , 1
while count <= n*n :
    if up <= down :
        for i in range(left,length-right) :
            matrix[up][i] = count
            count += 1
        up += 1
        for i in range(up,height-down) :
            matrix[i][length-right-1] = count
            count += 1
        right += 1
    else :
        for i in range(length-right-1,left-1,-1) :
            matrix[height-down-1][i] = count
            count += 1
        down += 1
        for i in range(height-down-1,up-1,-1) :
            matrix[i][left] = count
            count += 1
        left += 1
print matrix
```
