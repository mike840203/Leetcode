```python
dict = {'10': 'X', '50': 'L', '1': 'I', '5': 'V', '100': 'C', '500': 'D', '1000': 'M'}
out = "__"
n = 3
rank = 0

for i in range(10) :
    if not n / pow(10,i) :
        rank = i
        break
print "rank : " + str(rank)
for i in range(1,(n / 1000)+1) :
    out += dict['1000']  
for i in range(rank,0,-1) :  
    n = n % pow(10,i)
    print i , n 
    set = 0
    thisround = pow(10,i-1)
    if ((n / thisround) + 1 ) % 5 == 0 :
        out = out + dict[str(thisround)] + dict[str(((n / thisround) + 1) * thisround)]
    elif ((n / thisround) + 1 ) >= 6 :
        out = out + dict[str(5*thisround)]
        set = 1
    if set == 1 :
        for i in range(1,(n / thisround)+1-(((n / thisround))/5)*5) :
            out += dict[str(thisround)]
    else :
        for i in range(1,(n / thisround)+1-(((n / thisround))/5)*5) :
            out += dict[str(thisround)]
print out
```
