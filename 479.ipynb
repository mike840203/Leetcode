```python
n = 4
m = 0
def check(num) :
    num = str(num)
    if num == num[::-1] :
        return True
if n == 1 : print 9
for i in xrange(pow(10,n),0,-1) :
    for j in xrange(i,0,-1) :
        temp = i*j
        if temp > pow(10,(2*n)-1) :
            if check(temp) :
                    m = max(m,temp)
        else :
            break
print m 
```
