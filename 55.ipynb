```python
from operator import attrgetter
class Interval(object):
    def __init__(self, s=0, e=0):
        self.start = s
        self.end = e
list = [[1,4],[1,4]]
# [[1,6],[8,10],[15,18]]
intervals = []
for element in list :
    intervals.append(Interval(element[0],element[1]))
intervals = sorted(intervals,key=attrgetter('start'))
count = 0
while(count<=len(intervals)-2) :
    if intervals[count].end >= intervals[count+1].end :
        del intervals[count+1]
    elif intervals[count].end >= intervals[count+1].start :
        intervals.insert(count+2,Interval(intervals[count].start,intervals[count+1].end))
        del intervals[count]
        del intervals[count]
    else :
        count = count + 1
out = []
for element in intervals :
    out.append([element.start,element.end])
print out
```
