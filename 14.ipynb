```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        out = ''
        length = 100
        if len(strs) > 0 :
            for i in range(len(strs)) :
                if len(strs[i]) < length :
                    length = len(strs[i])
            print length
            for i in range(length) :
                common = strs[0][i]
                index = 0
                for j in range(len(strs)) :
                    if common == strs[j][i] :
                        index = index + 1
                    else :
                        return out
                print index
                if index == len(strs) :
                    out = out + common
        return out
```
