```python
s = "ebcbbececabbacecbbcbe"
x = [s,s[::-1]] 
out = False
for i in x :
    head = 0
    tail = len(i) - 1
    times = 0
    temp = True
    while head < tail :
        if i[head] == i[tail] :
            head += 1 
            tail -= 1
        elif i[head+1] == i[tail] and times == 0 :
            head += 2
            tail -= 1
            times = 1
        elif i[head] == i[tail-1] and times == 0:
            head += 1
            tail -= 2
            times = 1
        else :
            temp = False
            break
    out = out | temp
print out
```
