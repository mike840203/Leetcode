```python
pattern = "abba"
str = "dog cat cat ll"
str = str.split(' ')
dict = {}
out = True
if len(pattern) != len(str) : print False
for i in range(len(pattern)) :
    if not pattern[i] in dict :
        if str[i] in dict.values() :
            out = False
        dict[pattern[i]] = str[i]
    elif dict[pattern[i]] != str[i] :
        out = False
print out
        
```
