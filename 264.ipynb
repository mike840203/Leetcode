```python
num = 232
def isugly(num) :
    if num == 1 : return True
    for i in [2,3,5] :
        if num % i == 0 :
            return isugly(num/i)
    return False
count = 0
i = 1
while count < num :
    if isugly(i) :
        count = count + 1 
        #print '第 : ' + str(count) + ' 數字 : ' + str(i)
    i = i + 1 
print i - 1
```
