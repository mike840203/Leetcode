```python
grid = [[1,3,1],[1,5,1],[4,2,1]]
vertical = len(grid) - 1
horizontal = len(grid[0]) - 1
def recur(v,h,res) :
    if v < 0 or h < 0 : return 
    res += grid[v][h]
    if v == 0 and h == 0 :
        return res 
    elif v == 0 :
        return recur(v,h-1,res)
    elif h == 0 :
        return recur(v-1,h,res)
    return min(recur(v-1,h,res),recur(v,h-1,res))
recur(vertical,horizontal,0)
```
