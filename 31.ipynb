```python
nums = [1,1,5]
left_axis = -1
right_axis = 0
for i in range(len(nums)-1) :
    if nums[i] < nums[i+1]  :
        left_axis = i
for j in range(len(nums)-1,left_axis,-1) :
    if nums[j] > nums[left_axis] :
        right_axis = j
        break
if left_axis != -1 :
    temp = nums[left_axis]
    nums[left_axis] = nums[right_axis]
    nums[right_axis] = temp
head , tail = left_axis + 1 , len(nums) - 1
while head <= tail :
    temp = nums[head]
    nums[head] = nums[tail]
    nums[tail] = temp
    head += 1
    tail -= 1
print nums
```
