```python
nums1 = [1,1,2,4,6,0,0,0]
nums2 = [2,3,7]
m = 5
n = 3
for i in range(m,m+n) :
    nums1[i] = nums2[i-m]
for i in range(len(nums1)-1) :
    for j in range(i+1,len(nums1)) :
        if nums1[i] > nums1[j] :
            temp = nums1[i]
            nums1[i] = nums1[j]
            nums1[j] = temp
print nums1
```
