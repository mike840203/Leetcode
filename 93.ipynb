```python
s = "1111"
def IP(string, level, remain, output) :
    if len(remain) > (4-level+1) * 3 : # impossible
        return 
    elif len(remain) <= 3 and int(remain) > 0 and int(remain[::-1]) % 10 != 0 :
        output.append(string + "." + remain)
        return
    elif len(remain) > 3 : 
        length = 4 if len(remain) >=3 else len(remain)
        for i in range(1,length+1) :
            dot = "." if level > 1 else ""
            if int(remain[:i]) <= 255 and int(remain[:i:-1]) % 10 != 0 or remain[:i] == "0"  :
                IP(string + dot + remain[:i], level+1 , remain[i:],output)
    return output
print IP("",1,s,[])
```
