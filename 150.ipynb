```python
tokens = ["4","-2","/","2","-3","-","-"]
listx = []
out = 0 
def transfer(num1,num2,operator) :
    if operator == '+' :
        out = num1 + num2
    elif operator == '-' :
        out = num1 - num2
    elif operator == '*' :
        out = num1 * num2
    elif operator == '/' :
        out = num1 / num2
        if out < 0 and num1 % num2 != 0 :
            out = out + 1
    return out
for element in tokens :
    if element.isalnum() or len(element) > 1 :
        listx.append(element)
    else :
        num1 = listx.pop()
        num2 = listx.pop()
        listx.append(transfer(int(num2),int(num1),element))
print listx.pop()
```
