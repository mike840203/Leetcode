```python
def cacul(x,times) :
    out = ''
    temp = 0
    x = str(x)
    for i in range(len(x)-1) :
        if x[i] != x[i+1] :
            temp = temp + 1
            out = out + str(temp) + str(x[i])
            temp = 0
        else :
            temp = temp + 1
    if temp == 0 :
        out = out + '1' + str(x[i+1])
    else :
        out = out + str(temp+1) + str(x[i+1])
    if times == 0 :
        return out
    else :
        return cacul(out,times-1)
n = 5
if n == 1 :
    print '1'
elif n == 2 :
    print '11'
elif n == 3 :
    print  '21'
else :
    print cacul(21,n-4)
```
