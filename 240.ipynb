```python
def binary(list,target) :
    head , tail = 0 , len(list) 
    out = False
    while head < tail :
        old_head , old_tail = head , tail
        if target > list[(head+tail)/2] :
            head , tail = (head+tail)/2 , tail
        elif target < list[(head+tail)/2] :
            head , tail = head , (head+tail)/2
        else :
            out =  True
            break
        if old_head == head and old_tail == tail :
            out = False
            break
    print (head+tail)/2
    return out
```
