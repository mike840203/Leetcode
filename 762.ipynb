```python
import timeit
start = timeit.default_timer()
L = 209495
R = 218772
out = 0
def isPrime(num) :
    out = True
    i = 2 
    if num < 2 :
        return False
    while i * i <= num :
        if num % i == 0 :
            out = False
        i = i + 1
    return out
for i in range(L,R+1) :
    count = 0
    for x in range(len(bin(i))) :
        if bin(i)[x] == '1' :
            count = count + 1
    if isPrime(count) :
        out = out + 1
print out
stop = timeit.default_timer()
print '花費時間 :' + str(round((stop - start)*1000,4)) + 'ms'
```
